## ratings.csv

### 1. Basic info

```python
# basic information
ratings_file_path = "../data/raw_data/ratings.csv"
ratings = pd.read_csv(ratings_file_path, header=0)

print("Data shape:", ratings.shape)
print("\nThe first 5 rows are:")
print(ratings.head())

print("\nThe last 5 rows are:")
print(ratings.tail())

print("\nThe data types are:")
print(ratings.dtypes)

print("\nThe descriptive statistics are:")
print(ratings.describe(include='all'))
```

```python
ratings["userId"] = ratings["userId"].astype('category')
ratings["movieId"] = ratings["movieId"].astype('category')
```

```python
ratings["userId"].dtype
```

```python
# change column timestamp to date
ratings['datetime'] = pd.to_datetime(ratings["timestamp"], unit='s')
ratings.head(5)
```

```python
ratings['date'] = ratings['datetime'].dt.date
ratings.head(5)
```

### 2. Check missing values

```python
print("Missing values:")
print(ratings.isnull().sum())
```

```python
# check duplicated value
ratings.duplicated().sum()
```

```python
ratings.nunique()
```

```python
ratings.to_csv('../data/processed_data/processed_ratings.csv', index=False, header=True)
```
